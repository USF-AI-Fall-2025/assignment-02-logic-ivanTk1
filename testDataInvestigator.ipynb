```python
import pandas as pd
import matplotlib.pyplot as plt
from DataInvestigator import DataInvestigator
```

```python
#load the data
df = pd.read_csv('gallstone.csv')
```

```python
#columns to compare
columns_to_compare = [
    'Height', 'Weight', 'Body Mass Index (BMI)', 'Total Body Fat Ratio (TBFR) (%)',
    'Lean Mass (LM) (%)', 'Body Protein Content (Protein) (%)', 'Visceral Fat Rating (VFR)',
    'Bone Mass (BM)', 'Muscle Mass (MM)', 'Obesity (%)', 'Total Fat Content (TFC)',
    'Visceral Fat Area (VFA)', 'Visceral Muscle Area (VMA) (Kg)', 'Hemoglobin (HGB)'
]
col_indices = [df.columns.get_loc(col) for col in columns_to_compare]
```

```python
#calculate averages for each gender 
avg_gender_0 = []
avg_gender_1 = []
for idx in col_indices:
    avg_0 = df[df['Gender'] == 0][df.columns[idx]].mean()
    avg_1 = df[df['Gender'] == 1][df.columns[idx]].mean()
    avg_gender_0.append(avg_0)
    avg_gender_1.append(avg_1)

```

```python
#plot
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(columns_to_compare))
ax.bar(x, avg_gender_0, width=0.4, label='Gender 0', align='center')
ax.bar([i + 0.4 for i in x], avg_gender_1, width=0.4, label='Gender 1', align='center')

ax.set_xticks([i + 0.2 for i in x])
ax.set_xticklabels(columns_to_compare, rotation=45, ha='right')
ax.set_ylabel('Average Value')
ax.set_title('Average Values by Gender (using DataInvestigator)')
ax.legend()
plt.tight_layout()
plt.show()
```
