```python
import pandas as pd
import matplotlib.pyplot as plt
from DataInvestigator import DataInvestigator
```

```python
#load the data
df = pd.read_csv('gallstone.csv')
```

```python
#columns to compare
columns_to_compare = [
    'Height', 'Weight'
]
#gets indexes of the columns to compare
col_indices = [df.columns.get_loc(col) for col in columns_to_compare]
```

```python
#calculate averages for each gender 
avg_gender_0 = []
avg_gender_1 = []
for idx in col_indices:
    avg_0 = df[df['Gender'] == 0][df.columns[idx]].mean()
    avg_1 = df[df['Gender'] == 1][df.columns[idx]].mean()
    avg_gender_0.append(avg_0)
    avg_gender_1.append(avg_1)

```

```python
#For Question 1 (part 2)
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(columns_to_compare))
ax.bar(x, avg_gender_0, width=0.4, label='Gender 0', align='center')
ax.bar([i + 0.4 for i in x], avg_gender_1, width=0.4, label='Gender 1', align='center')

ax.set_xticks([i + 0.2 for i in x])
ax.set_xticklabels(columns_to_compare, rotation=45, ha='right')
ax.set_ylabel('Average Value')
ax.set_title('Average Values by Gender (using DataInvestigator)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
#For Question 2 (part 2)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
#Histogram
df['Glucose'].hist(bins=30)
plt.title('Glucose Histogram')
plt.xlabel('Glucose')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
#Boxplot
df.boxplot(column='Glucose')
plt.title('Glucose Boxplot')
plt.ylabel('Glucose')

plt.tight_layout()
plt.show()

#print median and mean
median_glucose = df['Glucose'].median()
mean_glucose = df['Glucose'].mean()
print("Median Glucose:", median_glucose)
print("Mean Glucose:", mean_glucose)
```
